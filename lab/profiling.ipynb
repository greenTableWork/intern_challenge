```python
from pathlib import Path
import pstats

profile_dir = (Path.cwd() / ".." / "profile").resolve()
```

```python
# reload latest profile
profiles = sorted(profile_dir.glob("*.prof"), key=lambda path: path.stat().st_mtime)

if not profiles:
    raise FileNotFoundError(f"No .prof files found in {profile_dir}")

latest_profile = profiles[-1]
print(f"Latest profile: {latest_profile}")
```

```python
pstats.Stats(str(latest_profile)).sort_stats("cumulative").print_stats()
```
